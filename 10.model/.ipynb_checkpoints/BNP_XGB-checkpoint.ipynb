```python
# 変数マップを作成する
def create_feature_map(features):
    outfile = open('xgb.fmap', 'w')
    i = 0
    for feat in features:
        outfile.write('{0}\t{1}\tq\n'.format(i, feat))
        i = i + 1

    outfile.close()
```

### カラムの除外・データのappend

```python
target = d['target'] # 正解をリストに格納しておく
id_submit = d_submit['ID']

d_drop_ID = d.drop(['ID', 'target'], axis=1) # IDを削除
d_drop_ID_submit = d_submit.drop(['ID'], axis= 1)

# とりあえずv22を変数除外してやってみる
d_drop_ID = d_drop_ID.drop(['v22'], axis=1)
d_drop_ID_submit = d_drop_ID_submit.drop(['v22'], axis=1)
```

```python
# 説明変数からv22を除外したやつ
explanatory_colums2 = ['v1',	'v2',	'v3',	'v4',	'v5',	'v6',	'v7',	'v8',	'v9',	'v10',	'v11',	'v12',	'v13',	'v14',	'v15',	'v16',	'v17',	
                      'v18',	'v19',	'v20',	'v21',	'v23',	'v24',	'v25',	'v26',	'v27',	'v28',	'v29',	'v30',	'v31',	'v32',	'v33',	'v34',	
                      'v35',	'v36',	'v37',	'v38',	'v39',	'v40',	'v41',	'v42',	'v43',	'v44',	'v45',	'v46',	'v47',	'v48',	'v49',	'v50',	'v51',	
                      'v52',	'v53',	'v54',	'v55',	'v56',	'v57',	'v58',	'v59',	'v60',	'v61',	'v62',	'v63',	'v64',	'v65',	'v66',	'v67',	'v68',	
                      'v69',	'v70',	'v71',	'v72',	'v73',	'v74',	'v75',	'v76',	'v77',	'v78',	'v79',	'v80',	'v81',	'v82',	'v83',	'v84',	'v85',	
                      'v86',	'v87',	'v88',	'v89',	'v90',	'v91',	'v92',	'v93',	'v94',	'v95',	'v96',	'v97',	'v98',	'v99',	'v100',	'v101',	'v102',	
                      'v103',	'v104',	'v105',	'v106',	'v107',	'v108',	'v109',	'v110',	'v111',	'v112',	'v113',	'v114',	'v115',	'v116',	'v117',	'v118',	
                      'v119',	'v120',	'v121',	'v122',	'v123',	'v124',	'v125',	'v126',	'v127',	'v128',	'v129',	'v130',	'v131',]
```

```python
# 説明変数のfeature mapを作成
create_feature_map(explanatory_colums2)
```

```python
rowTrain = d_drop_ID.shape[0]
rowTest = d_drop_ID_submit.shape[0]
#train = train.append(test)
```

```python
# 学習データとサブミットデータを縦に結合
d_merge = d_drop_ID.append(d_drop_ID_submit)
```

### カテゴリを数値に変換

```python
# カテゴリ変数を数値に変換
#from sklearn import preprocessing 
#for f in d_merge.columns: 
    if d[f].dtype=='object': 
        lbl = preprocessing.LabelEncoder() 
        lbl.fit(list(d_merge[f].values)) 
        d_merge[f] = lbl.transform(list(d_merge[f].values))
```

### one-hotエンコーディング

```python
d_merge_dummy = pd.get_dummies(d_merge)
```

```python
# 改めて、学習とテスト用に分割
d_test  = d_merge_dummy[rowTrain: rowTrain + rowTest]
d_train = d_merge_dummy[0:rowTrain]
```

### 変数間の相関を確認

```python
# Compute the correlation matrix
corr = d_train.corr()
```

```python
# array型に変換
d_train_array = np.array(d_train)
d_train_array = d_train_array.astype(float)

d_test_array = np.array(d_test)
d_test_array = d_test_array.astype(float)
```

### データをsplit

```python
# Split the Learning Set
X_train, X_valid, y_train, y_valid = train_test_split(d_train_array, target, test_size=0.3, random_state=1)
X_train, X_eval, y_train, y_eval  = train_test_split(X_train, y_train, test_size=0.2, random_state=1)
```

### 学習する

```python
# parameters setting
# n_estimators and early_stopping_rounds should be increased
model = xgb.XGBClassifier(missing=np.nan, 
                        max_depth=7, 
                        n_estimators=300, 
                        learning_rate=0.05, 
                        subsample=1, 
                        colsample_bytree=0.9, 
                        seed=2100,
                        objective= 'binary:logistic')
```

```python
# fitting
model.fit(X_train, 
          y_train,  
          early_stopping_rounds=35,  
          eval_metric="logloss", 
          eval_set=[(X_eval, y_eval)] # dataset to use validation for early-stopping 
         )
```

```python
# save model to file
pickle.dump(model, open("./02.pickle/model2.pickle", "wb"))
```

```python
# load model from file
model = pickle.load( open('./02.pickle/model2.pickle', 'rb'))
```

## 重要変数

```python
# weight : 
# gain :
# cover :
importance = model.get_booster().get_score(fmap='01.fmap/xgb2.fmap', importance_type='gain')
importance = sorted(importance.items(), key=operator.itemgetter(1))

# dataframe化
df_importance = pd.DataFrame(importance, columns=['features', 'score'])
df_importance['score'] = df_importance['score'] / df_importance['score'].sum()
```

```python
# 重要度を可視化
plt.figure()
#df_importance.plot()
df_importance.plot(kind='barh', x='features', y='score', legend=False, figsize=(15, 20))
plt.title('XGBoost Feature Importance')
plt.xlabel('relative importance')
plt.gcf().savefig('03.importance/feature_importance_xgb.png')
```

## validationデータで検証

```python
# pred_probaのアウトプットには、1に分類される確率が配列の2つ目に格納される
pred = model.predict_proba(X_valid)[:, 1]
```

```python
from  sklearn.metrics import log_loss
logloss_valid = log_loss(y_valid, pred)
logloss_valid
```

```python
# ROC曲線用の変数作成
fpr, tpr, thresholds = mtr.roc_curve(y_valid.values, pred)
roc_auc = auc(fpr, tpr)
```

```python
# ROC曲線を描画
lw = 1
plt.figure(figsize=(4,3),dpi=150)
plt.plot(fpr, tpr, color = 'red', lw = lw, label='AUC = %0.2f' % roc_auc)
plt.plot([0, 1], [0, 1], color='grey', lw=lw, linestyle='--')
plt.xlim([0.0, 1.0])
plt.ylim([0.0, 1.05])

plt.title("ROC curve")
plt.xlabel("False Positive Rate")
plt.ylabel("True Positive Rate")
plt.legend(loc="lower right")
plt.show()
```

### Submissionデータ作成

```python
# pred_probaのアウトプットには、1に分類される確率が配列の2つ目に格納される
pred_submit = model.predict_proba(d_test_array)[:, 1]
```
